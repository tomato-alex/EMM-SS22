### Aufgabe 3.  
Ein Fitnessstudio bietet seinen Gästen am „Tag der offenen Tür“ Fruchtcocktails an. Aus Orangen- und Ananassaft sollen mindestens 12 Liter und höchstens 24 Liter eines neuen Fruchtcocktails gemischt werden. Verwendet werden sollen mindestens 12 Liter Orangensaft, aber höchstens 5 Liter Ananassaft. Die Mischung schmeckt nur, wenn die Menge des Ananassaftes mindestens 1/7 und höchstens 1/3 des Orangensaftes beträgt. Der Orangensaft kostet €1,80 und der Ananassaft €4,50 je Liter. Wie viele Liter Saft jeder Sorte müssen genommen werden, wenn die Kosten für den Einkauf möglichst gering gehalten werden sollen?

a) Stellen Sie das lineare Programm auf.

c) Lösen Sie das lineare Programm graphisch.

d) Lösen Sie das lineare Programm durch Schnitt der bindenden Nebenbedingungen und
   berechnen Sie den Zielfunktionswert.

e) Lösen Sie das duale LP Problem mit dem Simplex Algorithmus.

f) Vergleichen Sie beide Lösungen. Welche Zusammenhänge in Bezug auf Dualität können Sie feststellen?

```python
## Code ist von UE Kurs - Vizualisierung
import numpy as np
from matplotlib import pyplot as plt

def drawArrow(maxX, maxY, c):
    plt.arrow(0, 0, c[0], c[1], head_width=0.2, fc='k')

def drawLines(maxX, maxY, c):
    delta = 1
    x = np.arange(0, maxX, delta)
    y = np.arange(0, maxY, delta)
    X, Y = np.meshgrid(x, y)
    Z = c[0] * X + c[1] * Y
    CS = plt.contour(X, Y, Z, levels=np.linspace(0, 100, 10), cmap='coolwarm')
    plt.clabel(CS, inline=1, fontsize=12, fmt='%1.1f')

def fillRectangle(ax, x, y, width, height):
    ax.add_patch(plt.Rectangle((x, y), width, height, facecolor="green", alpha=0.3))

def drawInequality(inequality, ax, maxX, maxY):
    delta = 0.1
    x = np.arange(0, maxX, delta)
    y = np.arange(0, maxY, delta)
    for xi in x:
        for yi in y:
            if inequality(xi, yi):
                fillRectangle(ax, xi, yi, delta, delta)

def finishPlot(maxX, maxY):
    plt.xlim(0, maxX)
    plt.ylim(0, maxY)
    plt.xlabel('$x$')
    plt.ylabel('$Y$')
    plt.grid(True)
    plt.show()

maxX = 30
maxY = 10

c = np.array([1.8, 4.5])
fig = plt.figure(figsize = (6, 6))
ax = fig.add_subplot(1, 1, 1)

drawArrow(maxX, maxY, c)
drawLines(maxX, maxY, c)

inequality = lambda x, y : x + y <= 24
inequality2 = lambda x, y: x + y >= 12
ineq3 = lambda x, y: x + 0*y >= 12
ineq4 = lambda x, y: 0*x + y <= 5
ineq5 = lambda x, y: - x/7 + y >= 0
ineq6 = lambda x, y: -x/3 + y <= 0

drawInequality(inequality, ax, maxX, maxY)
drawInequality(inequality2, ax, maxX, maxY)
drawInequality(ineq3, ax, maxX, maxY)
drawInequality(ineq4, ax, maxX, maxY)
drawInequality(ineq5, ax, maxX, maxY)
drawInequality(ineq6, ax, maxX, maxY)
finishPlot(maxX, maxY)
```
